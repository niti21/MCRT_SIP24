```python
from numpy import random
import matplotlib.pyplot as plt
import numpy as np

def monte_carlo_radiative_transfer(N, incident_angle, sigma_e, sigma_s, filling_factor):
    # N: Number of photons to simulate
    # sigma_e: Extinction coefficient
    # sigma_s: Scattering coefficient
    # filling_factor: Filling factor of the medium

    # Calculate absorption coefficient
    sigma_a = sigma_e - sigma_s

    # Calculate single scattering albedo
    albedo = sigma_s / sigma_e

    # Initialize arrays to store positions and direction cosines
    x = np.zeros(N)
    y = np.zeros(N)
    z = np.zeros(N)
    mu_x = np.zeros(N)
    mu_y = np.zeros(N)
    mu_z = np.cos(np.deg2rad(incident_angle)) * np.ones(N)
    weights = np.ones(N)

    # Initialize optical path, transmission coefficient, and exit angles/energy arrays
    optical_path = np.zeros(N)
    transmission_coefficient = np.zeros(N)
    exit_angles = []
    exit_energy = []
    exit_energy_1 = []

    def p(g):
        return (((1 - (b ** 2)) / (1 + (2 * b * np.cos(np.deg2rad(g))) + (b ** 2))) ** (3 / 2))

    # Start the simulation
    for i in range(N):
        while weights[i] > 0:
            # Determine optical path and step size
            q = random.random()
            l = -np.log(q)
            s = l / sigma_e

            # Update optical path
            optical_path[i] += s

            # Update position
            x[i] += mu_x[i] * s
            y[i] += mu_y[i] * s
            z[i] += mu_z[i] * s

            if x[i] >= 30:
                x[i] -= 30

            if y[i] >= 30:
                y[i] -= 30

            # Absorption and scattering
            delta_W = weights[i] * (1 - albedo)
            weights[i] -= delta_W

            # Check if the photon scatters
            if random.random() < albedo:
                # Determine scattering angle using Henyey-Greenstein phase function
                b = 0.2  # Anisotropy factor (change as needed)
                phi = 2 * np.pi * random.random()

                dp_max = (1 + b) / ((1 - b) ** 2)
                for g in range(180):
                    dp = p(g) * 2 * np.pi * np.sin(np.deg2rad(g))
                    v2 = random.uniform(0, dp_max)
                    if v2 < dp:
                        theta = g
                        break

                costheta = np.cos(np.deg2rad(theta))
                sintheta = np.sqrt(1 - costheta ** 2)
                cosphi = np.cos(phi)
                sinphi = np.sin(phi)

                # Update direction cosines
                if mu_z[i] == 1:
                    mu_x_prime = sintheta * cosphi
                    mu_y_prime = sintheta * sinphi
                    mu_z_prime = costheta
                elif mu_z[i] == -1:
                    mu_x_prime = sintheta * cosphi
                    mu_y_prime = -sintheta * sinphi
                    mu_z_prime = -costheta
                else:
                    mu_x_prime = (sintheta / np.sqrt(1 - mu_z[i]**2)) * (mu_x[i] * mu_z[i] * cosphi - mu_y[i] * sinphi) + mu_x[i] * costheta
                    mu_y_prime = (sintheta / np.sqrt(1 - mu_z[i]**2)) * (mu_y[i] * mu_z[i] * cosphi + mu_x[i] * sinphi) + mu_y[i] * costheta
                    mu_z_prime = -sintheta * cosphi * np.sqrt(1 - mu_z[i]**2) + mu_z[i] * costheta

                # Update direction cosines
                mu_x[i] = mu_x_prime
                mu_y[i] = mu_y_prime
                mu_z[i] = mu_z_prime

            # Update transmission coefficient
            transmission_coefficient[i] = np.exp(-optical_path[i] * sigma_e)

            # Check if the photon has exited the medium
            if z[i] > 12:
                # Store exit angle and corresponding energy
                exit_angles.append(np.rad2deg(np.arccos(mu_z[i])))
                exit_energy.append(transmission_coefficient[i])
                exit_energy_1.append((transmission_coefficient[i] * weights[i]) * (np.cos(np.deg2rad(np.rad2deg(np.arccos(mu_z[i])))) / np.cos(np.deg2rad(incident_angle))))
                break

    return exit_angles, exit_energy, exit_energy_1

# Calculate reflectance vs emission angle
def calculate_reflectance_vs_emission(exit_angles, exit_energy, incident_angle):
    reflectance_vs_emission = []
    for i in range(len(exit_angles)):
        r_mu_mu0 = (exit_energy_1[i] / exit_energy[i])
        reflectance_vs_emission.append((exit_angles[i], r_mu_mu0))
    return reflectance_vs_emission


# Example usage
N = 1000  # Number of photons to simulate
# Incident angle
incident_angle = 30 # in degrees
sigma_e = 0.1  # Extinction coefficient
sigma_s = 0.1  # Scattering coefficient
filling_factor = 0.2  # Filling factor

exit_angles, exit_energy, exit_energy_1 = monte_carlo_radiative_transfer(N, incident_angle, sigma_e, sigma_s, filling_factor)
#print(exit_angles)
reflectance_vs_emission = calculate_reflectance_vs_emission(exit_angles, exit_energy, incident_angle)

# Plotting reflectance vs emission angle
emission_angles = [item[0] for item in reflectance_vs_emission]
reflectance_values = [item[1] for item in reflectance_vs_emission]

plt.plot(emission_angles, reflectance_values, color='blue')
plt.xlabel('Emission Angle (degrees)')
plt.ylabel(r'$r \mu / \mu_0$')
plt.title(r'$r \mu / \mu_0$ vs Emission Angle')
plt.grid(True)
plt.show()

```
